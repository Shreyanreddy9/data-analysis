```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = "MI6_GW_STW_AP (1).csv"  # Update with correct path
df = pd.read_csv(file_path)

# Set visualization style
sns.set_style("whitegrid")

# Generate a set of plots
plots = {}
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Ensure seaborn theme is set
sns.set_theme(style="whitegrid")

# Create figure
plt.figure(figsize=(12, 6))

# Count occurrences of each scheme type
scheme_type_counts = df["scheme_type_name"].value_counts()

# Create bar plot
sns.barplot(x=scheme_type_counts.index, y=scheme_type_counts.values, palette="viridis")

# Rotate x-axis labels for readability
plt.xticks(rotation=90)

# Set plot labels and title
plt.title("Distribution of Scheme Types")
plt.xlabel("Scheme Type")
plt.ylabel("Count")

# Ensure plots dictionary is defined
plots = {}  # Define if not already
plots["scheme_type_distribution"] = plt.gcf()  # Store figure in dictionary
```

```python
plt.figure(figsize=(12, 6))
tubewell_counts = df["tubewell_type"].value_counts()
sns.barplot(x=tubewell_counts.index, y=tubewell_counts.values, palette="coolwarm")
plt.xticks(rotation=90)
plt.title("Distribution of Tubewell Types")
plt.xlabel("Tubewell Type")
plt.ylabel("Count")
plots["tubewell_type_distribution"] = plt.gcf()
```

```python
# Plot 3: Scheme Depth Distribution
plt.figure(figsize=(10, 5))
sns.histplot(df["scheme_depth"].dropna(), bins=30, kde=True, color="blue")
plt.title("Distribution of Scheme Depth")
plt.xlabel("Depth (m)")
plt.ylabel("Frequency")
plots["scheme_depth_distribution"] = plt.gcf()
```

```python
# Plot 4: Correlation Heatmap of Numerical Features
plt.figure(figsize=(14, 10))
numeric_df = df.select_dtypes(include=["float64", "int64"]).dropna(axis=1, how="all")
corr_matrix = numeric_df.corr()
sns.heatmap(corr_matrix, cmap="coolwarm", annot=False, linewidths=0.5)
plt.title("Correlation Heatmap of Numerical Features")
plots["correlation_heatmap"] = plt.gcf()
```

```python
# Plot 5: Scheme Construction Cost Distribution
plt.figure(figsize=(10, 5))
sns.histplot(df["scheme_construction_cost"].dropna(), bins=50, kde=True, color="green")
plt.title("Distribution of Scheme Construction Cost")
plt.xlabel("Cost")
plt.ylabel("Frequency")
plots["scheme_construction_cost_distribution"] = plt.gcf()

# Show plots
plt.show()
```

```python
import matplotlib.pyplot as plt

# Function to create pie charts
def create_pie_chart(column, title, top_n=5):
    plt.figure(figsize=(8, 8))
    data_counts = df[column].value_counts().nlargest(top_n)  # Select top N categories
    plt.pie(data_counts, labels=data_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette("pastel"))
    plt.title(title)
    plt.show()
```

```python
# Pie Chart 1: Scheme Type Distribution
create_pie_chart("scheme_type_name", "Distribution of Scheme Types")
```

```python
# Pie Chart 2: Tubewell Type Distribution
create_pie_chart("tubewell_type", "Distribution of Tubewell Types")
```

```python
# Pie Chart 3: Water Distribution Methods
create_pie_chart("water_distribution_method_name", "Water Distribution Methods")
```

```python
# Pie Chart 4: Finance Sources Distribution
create_pie_chart("finance_source_name1", "Finance Source Distribution")
```

```python
# Pie Chart 5: Gender-Based Ownership
create_pie_chart("gender_name", "Ownership by Gender")
```
