```python
import pandas as pd
# Load the Excel file
file_path = 'MIDMARKS.xlsx' # Replace with your file path
df = pd.read_excel(file_path)
# Specify the columns to sum up
columns_to_sum = ["DV", "M-II", "PP", "BEEE", "FL", "FIMS"]
# Convert columns to numeric, coercing errors to NaN
df[columns_to_sum] = df[columns_to_sum].apply(pd.to_numeric, errors='coerce')
# Create a new column "Total Value" by summing up the specified columns
df["Total Value"] = df[columns_to_sum].sum(axis=1)
# Save the updated DataFrame back to an Excel file
output_file_path = 'MIDMARKS_WITH_TOTAL.xlsx' # Replace with your desired outpu
df.to_excel(output_file_path, index=False)
print(f"File with 'Total Value' column saved to {output_file_path}")
```

```python
df
```

```python
def assign_grade(total):
    if total >= 90:
        return 'A+'
    elif total >= 80:
        return 'A'
    elif total >= 70:
        return 'B+'
    elif total >= 60:
        return 'B'
    elif total >= 50:
        return 'C'
    else:
        return 'F'
```

```python
df["Grade"] = df["Total Value"].apply(assign_grade)
output_file_path = 'MIDMARKS_WITH_GRADES.xlsx'
df.to_excel(output_file_path, index=False)
df
```

```python
max_total_marks = 600
df["Percentage"] = (df["Total Value"] / max_total_marks) * 100
df["Percentage"] = df["Percentage"].round()
df.to_excel(output_file_path, index=False)
df
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
data = pd.read_excel("MIDMARKS.XLSX")

# Specify subjects
subjects = ["DV"]

# Drop rows with missing values
data.dropna(inplace=True)

# Convert subject scores to numeric and handle NaN values
for subject in subjects:
    data[subject] = pd.to_numeric(data[subject], errors='coerce').fillna(0)

# Define the grade assignment function
def assign_grade(percentage):
    if 18 <= percentage <= 20:
        return 'Very Good'
    elif 13 <= percentage <= 14:
        return 'Average'
    elif 13 <= percentage <= 17:
        return 'Good'
    else:
        return 'Poor'

# Apply the function to assign grades
data['Programming_skills'] = data['DV'].apply(assign_grade)

# Count the number of students in each grade category
grade_counts = data['Programming_skills'].value_counts()

# Plot the pie chart
plt.pie(grade_counts, labels=grade_counts.index, autopct='%1.1f%%', colors=plt.cm.Paired.colors)
plt.title("Distribution of Programming Skills Grades")
plt.show()

# Display the data and the count
print(data)
```
