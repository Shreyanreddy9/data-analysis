```python
import seaborn as sns 
import pandas as pd
import matplotlib.pyplot as plt
average_scores=df[['Q1','Q2','Q3','Q4','Q5','Q6']].mean()
average_scores=df.plot(kind='bar',color='skyblue',edgecolor='black')
plt.title("Average scores per question")
plt.xlabel("Questions")
plt.ylabel("Average score")
plt.xtitle(rotation=0)
plt.show()
```

```python
dv["Q1"]=dv["Q1aM4"]+dv["Q1bM6"]
dv["Q2"]=dv["Q2aM6"]+dv["Q2bM4"]
dv["Q3"]=dv["Q3aM5"]+dv["Q3bM5"]
dv["Q4"]=dv["Q4aM3"]+dv["Q4bM7"]
dv["Q5"]=dv["Q5M10"]
dv["Q6"]=dv["Q6aM4"]+dv["Q6bM6"]
```

```python
dv.drop("Q1aM4",axis=1,inplace=True)
dv.drop("Q1bM6",axis=1,inplace=True)
dv.drop("Q2aM6",axis=1,inplace=True)
dv.drop("Q2bM4",axis=1,inplace=True)
dv.drop("Q3aM5",axis=1,inplace=True)
dv.drop("Q3bM5",axis=1,inplace=True)
dv.drop("Q4aM3",axis=1,inplace=True)
dv.drop("Q4bM7",axis=1,inplace=True)
dv.drop("Q5M10",axis=1,inplace=True)
dv.drop("Q6aM4",axis=1,inplace=True)
dv.drop("Q6bM6",axis=1,inplace=True)
```

```python
dv.fillna(0)
```

```python
questions = ["Q1", "Q2", "Q3", "Q4", "Q5", "Q6"]
marks = dv[questions].mean(axis=0).values
plt.figure(figsize=(10, 6))
sns.lineplot(x=questions, y=marks, marker="o", color="blue")
plt.title("Line Plot of Average Marks per Question", fontsize=16)
plt.xlabel("Questions", fontsize=12)
plt.ylabel("Average Marks", fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=questions, y=marks, color="red", s=50)
plt.title("Scatter Plot of Average Marks per Question", fontsize=16)
plt.xlabel("Questions", fontsize=12)
plt.ylabel("Average Marks", fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=dv[questions], palette="pastel")
plt.title("Box Plot of Question Scores", fontsize=16)
plt.xlabel("Questions", fontsize=12)
plt.ylabel("Marks", fontsize=12)
plt.show()
```
